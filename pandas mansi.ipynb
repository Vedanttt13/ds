```python
#1.Load the Dataset
import numpy as np

data_array = np.genfromtxt('/content/WorldCups1.csv', delimiter=',', skip_header=1, dtype=None, encoding=None)

print("Data Array:\n", data_array)
```

```python
# 2. Find the Average Attendance
attendance = [float(row[8]) for row in data_array] # Extract the 9th element (index 8) from each tuple as attendance
average_attendance = np.mean(attendance)
print("Average Attendance:", average_attendance)
```

```python
# 3. Find the Year with the Maximum Attendance
max_attendance_index = np.argmax(attendance)
year_max_attendance = data_array[max_attendance_index][0]
print("Year with Maximum Attendance:", year_max_attendance)
```

```python
# 4.Find the Unique Countries
countries = np.unique([row[1] for row in data_array])
print("Unique Countries:", countries)
```

```python
# 5. Find the Total Attendance
total_attendance = np.sum(attendance)
print("Total Attendance:", total_attendance)
```

```python
# 6. Find the Most Frequent Number of Qualified Teams
# The data_array was reshaped to have a single column, so we need to access index 0
unique_teams, counts = np.unique(data_array[:, 0], return_counts=True)
most_frequent_teams = unique_teams[np.argmax(counts)]
print("Most Frequent Number of Qualified Teams:", most_frequent_teams)
```

```python
# 7.Calculate the range of attendance
max_attendance = np.max(attendance)
min_attendance = np.min(attendance)
range_attendance = max_attendance - min_attendance
print("Range of Attendance:", range_attendance)
```

```python
# 8.Calculate the total number of matches played
total_matches = np.sum(data_array['f0'].astype(int))
print("Total Matches Played:", total_matches)
```

```python
# 9.Calculate the average number of qualified teams
avg_qualified_teams = np.mean(data_array['f0'].astype(int))
print("Average Number of Qualified Teams:", avg_qualified_teams)
```

```python
# 10.Find the most recent tournament year
most_recent_year = np.max(data_array['f0'])
print("Most Recent Tournament Year:", most_recent_year)
```

```python
# 11. Find the year with the maximum matches played
matches_played = data_array['f0'].astype(int)
max_matches_index = np.argmax(matches_played)
year_max_matches = data_array[max_matches_index, 0]
print("Year with Maximum Matches Played:", year_max_matches)
```

```python
# 12.Calculate the standard deviation of matches played
std_dev_matches = np.std(matches_played)
print("Standard Deviation of Matches Played:", std_dev_matches)
```

```python
# 13.Calculate the median number of qualified teams
qualified_teams = data_array['f0'].astype(int)  # Access the first column
median_qualified_teams = np.median(qualified_teams)
print("Median Number of Qualified Teams:", median_qualified_teams)
```

```python
# 14.Find the minimum attendance
min_attendance = np.min(attendance)
print("Minimum Attendance:", min_attendance)
```

```python
#15.Count the number of tournaments held in the 2000s
years = data_array[:, 0]['f0']

try:
    years = years.astype(int)
except ValueError:
    print("Error: Could not convert years to integers. Check for non-numeric values.")
    print("Unique values in the 'years' column:", np.unique(years))
else:
    count_2000s = np.sum((years >= 2000) & (years < 2010))
    print("Number of Tournaments Held in the 2000s:", count_2000s)
```

```python
# 16. Find the year with the minimum number of qualified teams
qualified_teams = data_array['f0']

try:
    qualified_teams_int = qualified_teams.astype(int)
except ValueError:
    print("Error: Could not convert qualified teams to integers. Check for non-numeric values.")
    print("Unique values in 'qualified_teams':", np.unique(qualified_teams))
else:
    min_qualified_index = np.argmin(qualified_teams_int)
    year_min_qualified = data_array[min_qualified_index, 0]
    print("Year with Minimum Number of Qualified Teams:", year_min_qualified)
```

```python
# 17.Determine the most common winner
winners = data_array[:, 0]['f2']
unique_winners, counts = np.unique(winners, return_counts=True)
most_common_winner = unique_winners[np.argmax(counts)]
print("Most Common Winner:", most_common_winner)
```

```python
# 18. Find the minimum number of matches played
# Assuming the number of matches is in the 1st column (index 0)
min_matches = np.min(data_array[:, 0]['f7'].astype(int))
print("Minimum Matches Played:", min_matches)
```

```python
# 19.Count the number of tournaments with more than 20 matches played
count_more_than_20_matches = np.sum(data_array[:, 0]['f7'].astype (int) > 20)
print("Number of Tournaments with More Than 20 Matches Played:", count_more_than_20_matches)
```

```python
# 20.Count the number of tournaments held in the 1980s
years = data_array[:, 0]['f0'].astype(int)  # Access the 'f0' field for years
count_1980s = np.sum((years >= 1980) & (years < 1990))
print("Number of Tournaments Held in the 1980s:", count_1980s)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/WorldCups1.csv')
```

```python
#1)How has the number of goals scored per tournament changed over time?
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['GoalsScored'], marker='o')
plt.title('Goals Scored Over Time')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.grid(True)
plt.show()
```

```python
#2) How has the number of qualified teams participating in the world Cup changed over time?
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['QualifiedTeams'], marker= 'o', color="orange")
plt.title('Number of Qualified Teams Over Time')
plt.xlabel('Year')
plt.ylabel("Qualified Teams")
plt.grid(True)
plt.show()
```

```python
#3)Is there any correlation between the attendance and the number of goals scored in World Cup tournaments?
plt.figure(figsize=(10, 6))
plt.scatter(df['Attendance'], df[ 'GoalsScored'], color="magenta")
plt.title("Attendance vs. Goals Scored in World Cup Tournaments")
plt.xlabel('Attendance')
plt.ylabel('Goals Scored')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4) How many goals were scored by the host country in each World Cup?
plt.figure(figsize=(12, 6))
plt.bar(df['Year'].astype(str), df['GoalsScored'], color='darkblue')
plt.title('Goals Scored by Host Country in Each World Cup')
plt.xlabel('Year')
plt.ylabel('Goals Scored')
plt.xticks (rotation=45)
plt.tight_layout()
plt.show()
```

```python
#5) which top 5 countries have scored the most goals in World Cup history?
top_countries = df[['Winner', 'GoalsScored']].groupby('Winner').sum().nlargest (5, 'GoalsScored')
plt.figure(figsize=(10, 6))
plt.bar(top_countries.index, top_countries['GoalsScored'], color='purple')
plt.title('Top 5 Countries by Goals Scored in World Cup History')
plt.xlabel('Country')
plt.ylabel('Total Goals Scored')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#6) How do the total goals scored relate to the number of teams participating in each tournament?
plt.figure(figsize=(10, 6))
plt.scatter(df['QualifiedTeams'], df['GoalsScored'], color='teal')
plt.title('Goals Scored vs. Number of Qualified Teams')
plt.xlabel('Number of Qualified Teams')
plt.ylabel('Goals Scored')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#7) How do the top-scoring teams performances in terms of goals scored compare across different tournaments?
top_teams = df[['Winner', 'GoalsScored']].groupby('Winner').sum().nlargest (10,'GoalsScored')
plt.figure(figsize=(12, 6))
top_teams['GoalsScored'].plot(kind='bar', color="darkgreen")
plt.title("Performance of Top Scoring Teams in World Cup History")
plt.xlabel("Team")
plt.ylabel('Total Goals Scored')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#8)Is there a relationship between the number of goals scored and the number of matches played in each World Cup?
plt.figure(figsize=(10, 6))
plt.scatter(df["MatchesPlayed"], df[ 'GoalsScored'], color="blue")
plt.title('Goals Scored vs. Matches Played in World Cup Tournaments')
plt.xlabel('Matches Played')
plt.ylabel('Goals Scored')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#9)Goals Scored in the Final Match vs. Total Goals
df['FinalMatchGoals'] = df['GoalsScored']* 0.1 # Example adjustment
plt.figure(figsize=(10, 6))
plt.scatter(df ['GoalsScored'], df['FinalMatchGoals'], color='orange')
plt.title('Final Match Goals vs. Total Goals Scored')
plt.xlabel('Total Goals Scored')
plt.ylabel('Final Match Goals')
plt.xscale('log') # Use log scale for better visualization
plt.yscale('log') # Use log scale for better visualization
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#10) How do cumulative goals scored accumulate over the years?
df['CumulativeGoals'] = df['GoalsScored'].cumsum()
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['CumulativeGoals'], marker='o', linestyle='-', color="brown")
plt.title('Cumulative Goals Scored by Year')
plt.xlabel('Year')
plt.ylabel('Cumulative Goals Scored')
plt.grid(True)
plt.xticks (rotation=45)
plt.tight_layout()
plt.show()
```

```python
#11.Plot a Bar Chart of the Number of Tournaments Won by Each Country
win_counts= df['Winner'].value_counts()
plt.figure(figsize=(10, 6))
win_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.title('Number of Tournaments Won by Each Country')
plt.xticks(rotation=45)
plt.show()
```

```python
#12. create a pie chart of distribution of goals scored in different decdes
df['Decade'] = (df['Year'] // 10) * 10
goals_by_decade = df.groupby('Decade')['GoalsScored'].sum()
plt.figure(figsize=(10, 6))
goals_by_decade.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Goals Scored in Different Decades')
plt.show()
```

```python
#13.create a scatter plot of goals scored vs,attendance
plt.figure(figsize=(10, 6))
plt.scatter(df['GoalsScored'], df['Attendance'], color='red')
plt.title('Goals Scored vs. Attendance')
plt.xlabel('Goals Scored')
plt.ylabel('Attendance')
```

```python
#14.plot a stacked bar chart of goals scored by country
country_goals = df.groupby('Winner')['GoalsScored'].sum()
plt.figure(figsize=(10, 6))
country_goals.plot(kind='bar', stacked=True)
plt.title('Goals Scored by Country')
```

```python
#15. plot a simple bar chart of number of tournaments per year
tournaments_per_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
tournaments_per_year.plot(kind='bar')
plt.title('Number of Tournaments per Year')
```

```python
#16.create a pie chart showing the proportion of goals scored by each country
goals_by_country = df['Winner'].value_counts()
plt.figure(figsize=(10, 6))
goals_by_country.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Goals Scored by Country')
plt.show()
```
