```python
# Import necessary libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules


# Sample data loading (replace with your dataset path)
data = pd.read_csv('/content/OnlineRetail.csv', encoding='ISO-8859-1')

# Preprocessing
data = data[data['Quantity'] > 0]  # Filter for positive quantities
data['Description'] = data['Description'].fillna('Unknown')  # Handle missing descriptions
data['Description'] = data['Description'].astype(str)  # Ensure descriptions are strings

# Group products by InvoiceNo to form transactions
transactions = data.groupby('InvoiceNo')['Description'].apply(list).values.tolist()

# Transaction encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)  # One-hot encoding
df = pd.DataFrame(te_ary, columns=te.columns_)  # Convert to DataFrame

# Apriori Algorithm - Frequent Itemsets Generation

frequent_itemsets_apriori = apriori(df, min_support=0.02, use_colnames=True)  # Minimum support = 2%


# Association Rules Generation
rules_apriori = association_rules(
    frequent_itemsets_apriori,
    metric="lift",
    min_threshold=1.0,
    num_itemsets=len(frequent_itemsets_apriori)
)

# Output Results
print(f"Frequent Itemsets (Apriori):\n", frequent_itemsets_apriori.head())
print(f"Association Rules (Apriori):\n", rules_apriori.head())
print(f"Execution Time for Apriori: {apriori_time:.2f} seconds")








!pip install apyori
from apyori import apriori

import pandas as pd


 ds = pd.read_csv("Market_Basket.csv", header=None)
ds.fillna(0, inplace=True)\

len(ds)
ds.columns

transaction = []
for i in range(0, 7501):
  transaction.append([str(ds.values[i,j]) for j in range(0, 20)])
print(transaction)


rules = apriori(transactions = transaction, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)

rules
Result = list(rules)
Result

df_result = pd.DataFrame(Result)

df_result.head()

  for item in Result:
    pair = item[0]
    items = [x for x in pair]
    print("Rule : " + items[0] + " - >" + items[1])
    print("confidence: " + str(item[1]))
    print("Lift: " + str(item[2][0][3]))
    print("==============================================")
```

```python
# Import necessary libraries
import pandas as pd
from efficient_apriori import apriori

# Load the dataset
ds = pd.read_csv("/content/Market_Basket.csv", header=None)

# Preprocessing
ds.fillna(0, inplace=True)

# Convert data into transactions list
transactions = []
for i in range(0, 7501):
    transaction = [str(ds.values[i, j]) for j in range(0, 20)]
    # Only keep non-zero items in the transaction
    transaction = [item for item in transaction if item != '0']
    transactions.append(transaction)

# Running the Eclat algorithm (Eclat-like via efficient-apriori)
# Removed min_lift parameter since it's not supported in efficient-apriori
itemsets, rules = apriori(transactions, min_support=0.003, min_confidence=0.2)

# Display the frequent itemsets and association rules
print(f"Frequent Itemsets (Eclat-like):\n", list(itemsets.items())[:5])  # Top 5 frequent itemsets (convert to list)
print(f"Association Rules (Eclat-like):\n", rules[:5])  # Top 5 rules

# Calculate and print support, confidence, and lift for rules
for rule in rules[:5]:
    print(f"Rule: {rule}, Support: {rule.support}, Confidence: {rule.confidence}, Lift: {rule.lift}")


```
