```python
# 3. KDE Plot

plt.figure(figsize=(12, 8))
sns.kdeplot(df['votes'], shade=True)
plt.title('KDE Plot of Votes')
plt.xlabel('Votes')
plt.show()
```

```python
# 4. Stack Plot

df_grouped = df.groupby('year')['votes'].sum().reset_index()
years = df_grouped['year'].values
votes = df_grouped['votes'].values
plt.figure(figsize=(12, 8))
plt.stackplot(years, votes, labels=['Votes'])
plt.title('Stack Plot of Votes by Year')
plt.xlabel('Year')
plt.ylabel('Votes')
plt.legend(loc='upper left')
plt.show()
```

```python
# 5. Polar Plot

plt.figure(figsize=(8, 8))
angles = np.linspace(0, 2 * np.pi, len(df_grouped['year']), endpoint=False)
votes_normalized = votes / max(votes)
angles = np.concatenate((angles, [angles[0]]))
votes_normalized = np.concatenate((votes_normalized, [votes_normalized[0]]))
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.plot(angles, votes_normalized, 'o-', linewidth=2)
ax.fill(angles, votes_normalized, alpha=0.25)
ax.set_yticklabels([])
ax.set_title('Polar Plot of Votes by Year', size=20)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(df_grouped['year'].astype(int), size=10)
plt.show()
```

```python
# 6. Contour Plot

x = np.linspace(min(df['year']), max(df['year']), 100)
y = np.linspace(min(df['votes']), max(df['votes']), 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(12, 8))
plt.contourf(X, Y, Z, cmap='RdBu')
plt.colorbar()
plt.title('Contour Plot of Generated Data')
plt.xlabel('Year')
plt.ylabel('Votes')
plt.show()
```

```python

df_pareto = df[['state', 'votes']].groupby('state').sum().sort_values('votes', ascending=False)
df_pareto['cumperc'] = df_pareto['votes'].cumsum() / df_pareto['votes'].sum() * 100

fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(df_pareto.index, df_pareto['votes'], color='C0')
ax2 = ax.twinx()
ax2.plot(df_pareto.index, df_pareto['cumperc'], color='C1', marker='D', ms=7)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}%'))
ax.tick_params(axis='x', rotation=90)
ax.set_title('Pareto Chart of Votes by State')
ax.set_xlabel('State')
ax.set_ylabel('Votes')
plt.show()
```

```python
# 8. Donut Chart
# ------------------------------
plt.figure(figsize=(10, 10))
df_state_votes = df['state'].value_counts().reset_index()
df_state_votes.columns = ['state', 'votes']
plt.pie(df_state_votes['votes'], labels=df_state_votes['state'], autopct='%1.1f%%', startangle=140, pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart of State Distribution')
plt.show()
```

```python
import matplotlib.ticker as ticker

# Group and sort data for the Pareto Chart
df_pareto = df[['state', 'votes']].groupby('state').sum().sort_values('votes', ascending=False)
df_pareto['cumperc'] = df_pareto['votes'].cumsum() / df_pareto['votes'].sum() * 100

# Limit to top 10 states (optional)
top_n = 10
df_pareto = df_pareto.head(top_n)

# Create the Pareto Chart
fig, ax = plt.subplots(figsize=(12, 8))

# Bar plot for votes by state
ax.bar(df_pareto.index, df_pareto['votes'], color='C0')

# Line plot for cumulative percentage
ax2 = ax.twinx()
ax2.plot(df_pareto.index, df_pareto['cumperc'], color='C1', marker='D', ms=7)
ax2.yaxis.set_major_formatter(ticker.PercentFormatter())

# Adjustments for readability
ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability
ax.set_title('Pareto Chart of Votes by State (Top 10)')
ax.set_xlabel('State')
ax.set_ylabel('Votes')
ax2.set_ylabel('Cumulative Percentage')

plt.show()
```
